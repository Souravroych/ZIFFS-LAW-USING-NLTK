```python
import nltk
from nltk import *
from nltk.corpus import gutenberg
from nltk.corpus import stopwords
from string import punctuation
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.tokenize import RegexpTokenizer
import numpy as np
import re
import string
import pylab
import matplotlib.pyplot as pyplt
from nltk.text import Text
from collections import Counter

######IMPORTING WORDS FROM CORPUS#######
alice = nltk.corpus.gutenberg.words('carroll-alice.txt')

######DELETING PUNCTUATION########
def strip_punctuation(s):
    return [''.join(c for c in string if c not in punctuation) for string in alice]
#print (strip_punctuation(alice))

alice_strip = strip_punctuation(alice)

# alice_words = re.sub("[^\w]"," ",alice_strip).split(' ')
# print(alice_words)

# alice_tokenized = nltk.word_tokenize(alice_strip)
# alice_tokenized

###########UNDERCASING THE CORPUS
alice_lower = [words.lower() for words in alice_strip]


############PRINTING THE LENGTH OF CORPUS
length_alice = len(alice_lower)
print('The number of words in corpus:',length_alice)

########PRINTING UNIQUE WORDS


########FREQUENCY OF OCCURENCE
freq_alice = collections.Counter(alice_lower)
freq_alice = freq_alice.most_common()
print('Frequency of occurence of words: ',freq_alice)

```

```python
##########Printing top 25 words and their occurence
print("Top 25 words and their occurence")
#top25_alice = freq_alice.most_common(25)
count=0
for key in freq_alice:
    if count < 25:
        print(key[0],key[1])
        count +=1
    
###########Top 25 c words
print("\n\n Top 25 frequent starting with 'c' Words\n\n")
count = 0
for key in freq_alice:

    if count < 25:
        if key[0].startswith('c'):
            print(key[0],key[1])
            count += 1


```

```python
for key in freq_alice:
    if count < 25:
        if key[0].startswith('c'):
            print(key[0],key[1])
            count += 1
```

```python
rank = 0
number_of_words = 0
word_rank = {}
count = 9999
for key in freq_alice:
        if key[1] < count:
            word_rank[key[0]] = rank + 1
            rank += 1
            count  = key[1]
        elif key[1] == count:
            word_rank[key[0]] = rank
print(word_rank)            
```

```python
for key in freq_alice:
    print(key[0],key[1],key[1]/length_alice,(key[1]/length_alice)*word_rank[key[0]])
```

```python
list_plot = []
print(freq_alice)
for f in freq_alice:
    freq = f[1]
    list_plot.append(freq)

pylab.scatter([*word_rank.values()],list_plot)
pyplt.title('ziffs law')
pyplt.xlabel('RANK')
pyplt.ylabel('OCCURENCE OF WORDS IN CORPUS')
pylab.show()
```
